```python
#Import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
#Import data in CSV format

file_path = '/Users/cameronyoungblood/Desktop/KAG_conversion_data.csv'

df = pd.read_csv(file_path)
```

```python
#Top values in dataset

df.head(10)
```

```python
#Drop Approved_Conversions column

df.drop('Approved_Conversion', axis=1, inplace=True)
```

```python
#Check the number of null values in each column

df.isnull().sum()
```

```python
#Get general overview of dataframe

df.info()
```

```python
#Descriptive stats about the dataframe

df.describe()
```

```python
#Create and set index of dataset

index = np.arange(1, 1144)

df.set_index(index, inplace=True)
```

```python
#Graphic overview of each variable in the dataset using seaborn pairplot

sb.pairplot(df)
```

```python
#Check for correlation of the variables in a heat map

correlation = df.corr()

sb.heatmap(correlation, annot=True)
```

```python
#Looking at the targeted ages in the campaigns

df.age.value_counts().plot(kind='bar')
plt.xticks(rotation=45)
plt.xlabel('Ages')
plt.title('Ages Targeted in Campaigns') 
```

```python
#Looking at the targeted genders in the campaigns

df.gender.value_counts().plot(kind='bar')
plt.xticks(rotation=0)
plt.xlabel('Gender')
plt.title('Gender Targeting in Campaigns')
```
